# Tensores

## Lista Python para Tensor

```python
matriz = [ [1,3,4] , [2,3,4] ]

tns = torch.FloatTensor(matriz)
print(tns, tns.dtype)

tns = torch.DoubleTensor(matriz)
print(tns, tns.dtype)

tns = torch.LongTensor(matriz)
print(tns, tns.dtype)
```

# Camada Linear

```python
import torch
from torch import nn
```

```python
torch.manual_seed(42)

perceptron = nn.Linear( in_features = 3, out_features = 1 )
print(perceptron)
```

```python
for nome, tensor in perceptron.named_parameters():
    print(nome, tensor.data)
```

```python
print(perceptron.weight.data)
print(perceptron.bias.data)
```

```python
w1, w2, w3 = perceptron.weight.data.numpy()[0]
b = perceptron.bias.data.numpy()[0]

print(w1, w2, w3, b)
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot3d(perceptron):
    w1, w2, w3 = perceptron.weight.data.numpy()[0]
    b = perceptron.bias.data.numpy()[0]

    x = np.linspace(-1,1,10)
    y = np.linspace(-1,1,10)
    
    X1, X2 = np.meshgrid(x,y)
    X3 = (b - w1*X1 - w2*X2) / w3
    
    fig = plt.figure()
    ax = fig.add_subplot(projection = '3d')
    ax.view_init(azim = 180)
    
    surf = ax.plot_surface(X1, X2, X3, cmap = "plasma")

plot3d(perceptron)
```

## Foward

```python
X = torch.Tensor( [0, -1, 2] )
y = perceptron(X)
print(y)

plot3d(perceptron)
plt.plot(X[0], X[1], X[2], marker = "^", color = "r", markersize = 20)
plt.show()
```

## Funções de ativação

```python
from sklearn.datasets import make_classification

np.random.seed(46)

X, Y = make_classification( n_features = 2, n_redundant = 0, n_informative = 1, n_clusters_per_class = 1 )
```

```python
def plotmodel(w1, w2, b):
    plt.scatter( X[:, 0], X[:, 1], marker = "o", c = Y, edgecolor = "k" )

    xmin, xmax = plt.gca().get_xlim()
    ymin, ymax = plt.gca().get_ylim()

    x = np.linspace( -2, 4, 50 )
    y = (-w1 * x -b ) / w2

    plt.axvline( 0, -1, 1, color = "k" )
    plt.axhline( 0, -2, 4, color = "k" )
    plt.plot(x, y, label = "_nolegend_")

    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
```

```python
w1 = 5
w2 = 1
b = 1.2
plotmodel(w1, w2, b)
```

```python
perceptron = nn.Linear(2, 1)
sigmoide = nn.Sigmoid()

perceptron.weight = nn.Parameter(torch.Tensor([[w1, w2]]) )
perceptron.bias = nn.Parameter(torch.Tensor([b]) )

```

```python
markers = ["^", "v", ">", "<"]
colors = ["r", "g", "b", "gray"]

plt.figure( figsize = (8, 6) )
plotmodel(w1, w2, b)
for k, idx in enumerate( [17, 21, 43, 66] ):
    x = torch.Tensor( X[idx] )

    ret = perceptron(x)
    act = sigmoide(ret)

    act_limiar = 0 if ret.data < 0 else 1

    label = "ret: {:5.2f}".format(ret.data.numpy()[0]) + " limiar: {:4.2f}".format(act_limiar) + " act: {:5.2f}".format(act.data.numpy()[0])
    plt.plot(x[0], x[1], marker = markers[k], markersize = 10, c = colors[k], label = label)

plt.legend()
plt.show()
```
