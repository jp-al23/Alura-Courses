# Inicialização

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Criação do dataframe

```python
df = pd.read_csv("dados/imigrantes_canada.csv")

df.head()
```

```python
df.info()
```

```python
df.set_index("País", inplace = True)
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
df[anos].head()
```

```python
brasil = df.loc["Brasil", anos]
brasil.head()
```

```python
brasil_df = pd.DataFrame( {"Ano": brasil.index.tolist(), "Imigrantes" : brasil.values.tolist()} )
brasil_df.head()
```

# Imigrações do Brasil

```python
plt.figure( figsize = (8,4) )
plt.plot( brasil_df.Ano, brasil_df.Imigrantes )
plt.title("Imigração do Brasil para o Canadá")
plt.xlabel("Ano")
plt.ylabel("Número de imigrantes")
plt.xticks(["1980", "1985", "1990", "1995", "2000", "2005", "2010", "2015"])
plt.show()
```

# Desafio: Imigrações do Brasil e Argentina

```python
argentina = df.loc["Argentina", anos]
argentina_df = pd.DataFrame( {"Ano": argentina.index.tolist(), "Imigrantes" : argentina.values.tolist()} )
argentina_df.head()
```

```python
plt.figure( figsize = (8,4) )
plt.plot( brasil_df.Ano, brasil_df.Imigrantes , label = "Brasil")
plt.plot( argentina_df.Ano, argentina_df.Imigrantes , label = "Argentina")
plt.title("Imigração do Brasil/Argentina para o Canadá")
plt.xlabel("Ano")
plt.ylabel("Número de imigrantes")
plt.xticks(["1980", "1985", "1990", "1995", "2000", "2005", "2010", "2015"])
plt.legend()
plt.show()
```

# Criação de figuras

```python
fig, ax = plt.subplots( figsize = (8, 4) )
ax.plot( brasil_df.Ano, brasil_df.Imigrantes )
ax.set_title("Imigração do Brasil para o Canadá\n1980 a 2013")
ax.set_xlabel("Ano")
ax.set_ylabel("Número de imigrantes")
ax.xaxis.set_major_locator( plt.MultipleLocator(5) )
plt.show()
```

```python
fig, axs = plt.subplots( nrows = 1, ncols = 2 , figsize = (15,5) )

# Gráfico de linhas
axs[0].plot( brasil_df.Ano, brasil_df.Imigrantes )
axs[0].set_title("Imigração do Brasil para o Canadá\n1980 a 2013")
axs[0].set_xlabel("Ano")
axs[0].set_ylabel("Número de imigrantes")
axs[0].xaxis.set_major_locator( plt.MultipleLocator(5) )
axs[0].grid()

# Boxplot
axs[1].boxplot(brasil_df.Imigrantes)
axs[1].set_title("Boxplot da imigração do Brasil para o Canadá\n1980 a 2013")
axs[1].set_xlabel("Brasil")
axs[1].set_ylabel("Número de imigrantes")
axs[1].grid()


plt.show()
```

# Figura com 4 subplots

```python
fig, axs = plt.subplots(2, 2, figsize = (10,6), sharey = True)
fig.subplots_adjust( hspace = 0.5, wspace = 0.3 )
fig.suptitle("Imigração dos 4 maiores países da América do Sul para o Canadá de 1980 a 2013")

paises = ["Brasil", "Colômbia", "Peru", "Argentina"]
for i in range(4):
    linha = i//2
    coluna = i%2
    pais = paises[i]

    axs[linha,coluna].plot( df.loc[pais, anos] )
    axs[linha,coluna].set_title(pais)
    axs[linha,coluna].xaxis.set_major_locator( plt.MultipleLocator(5) )
    axs[linha,coluna].set_xlabel("Ano")
    axs[linha,coluna].set_ylabel("Número de imigrantes")
    axs[linha,coluna].grid()

plt.show()
```

# Desafio: Vendas de diferentes lojas

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

desafio2_df = pd.DataFrame( vendas_2022, index = lojas).T
desafio2_df.head()
```

```python
fig, axs = plt.subplots(2, 2, figsize = (10, 6))
fig.subplots_adjust( hspace = 0.4 )
fig.suptitle("Número de vendas mensal de 4 lojas no período de 2022")

for i, loja in enumerate(desafio2_df.columns):
    linha = i//2
    coluna = i%2
    axs[linha, coluna].plot( desafio2_df.index, desafio2_df[loja] )
    axs[linha, coluna].set_title("Loja " + loja)
    axs[linha, coluna].set_xlabel("Mês")
    axs[linha, coluna].set_ylabel("Número de vendas")
    axs[linha, coluna].grid()

ymin = 0
ymax = desafio2_df.max().max()

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```

# Alterando cores + Anotações

## Coletando os países da América do Sul

```python
america_do_sul = df.query("Região == 'América do Sul'").sort_values(by = "Total")
america_do_sul
```

## Gráfico

```python
cores = ["green" if pais == "Brasil" else "silver" for pais in america_do_sul.index]

fig, ax = plt.subplots( figsize = (12,5) )
ax.barh(america_do_sul.index, america_do_sul.Total, color = cores)
ax.set_title("América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013", loc = "left", fontsize = 18)
ax.set_ylabel("")
ax.set_xlabel("Número de imigrantes", fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
ax.grid(linestyle = "--", axis = "x")

# Comentários
for i, v in enumerate(america_do_sul.Total):
    ax.text( v + 25, i, str(v) , color = "black" , fontsize = 8, ha = "left", va = "center")
    # ha = alinhamento horizontal
    # va = alinhamento vertical

ax.set_frame_on(False) # Remove a caixinha
ax.get_xaxis().set_visible(False) # Remover o eixo x
ax.tick_params(axis = "both", which = "both", length = 0)

plt.show()
```
